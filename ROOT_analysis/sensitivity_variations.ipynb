```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
base_dir = "/home/lmlepin/Desktop/dm_sets/dark_tridents_analysis/sensitivity_files/"
plots_dir = "/home/lmlepin/Desktop/Plots_DT_Drive/2023/sensitivity_plots/"

df_run1 = pd.read_csv(base_dir + "fermion_bdt_sensitivity_alpha_0.1_run1.csv")
df_run3 = pd.read_csv(base_dir + "fermion_bdt_sensitivity_alpha_0.1_run3.csv")
df_all = pd.read_csv(base_dir + "fermion_bdt_sensitivity_alpha_0.1_all_runs.csv")
```

```python
plt.figure(figsize=(17,14),dpi=300)
plt.plot(df_run1['mass'],df_run1['epsilon_squared'],color='red',label='run1')
plt.plot(df_run3['mass'],df_run3['epsilon_squared'],color='blue',label='run3')
plt.plot(df_all['mass'], df_all['epsilon_squared'], color='black',label='run1 + run3')
plt.yscale('log')
plt.xscale('log')
plt.xticks(size=25)
plt.yticks(size=25)
plt.xlabel(r'$M_{A^{\prime}}$[GeV]',size=30, labelpad=20)
plt.ylabel(r'$\epsilon^2$',size=30, labelpad=20)
plt.minorticks_off()
plt.legend(fontsize=20)
plt.savefig(plots_dir + "sensitivity_per_run_contribution.png")
plt.show()
```

```python
df_08 = pd.read_csv(base_dir + "fermion_bdt_sensitivity_alpha_0.1_all_runs_0.8_bkg_uncert.csv")
df_05 = pd.read_csv(base_dir + "fermion_bdt_sensitivity_alpha_0.1_all_runs_0.5_bkg_uncert.csv")
df_01 = pd.read_csv(base_dir + "fermion_bdt_sensitivity_alpha_0.1_all_runs_0.1_bkg_uncert.csv")
plt.figure(figsize=(17,14),dpi=300)
plt.plot(df_08['mass'],df_08['epsilon_squared'],color='red',label='80% background uncertainty')
plt.plot(df_05['mass'],df_05['epsilon_squared'],color='blue',label='50% background uncertainty')
plt.plot(df_01['mass'],df_01['epsilon_squared'],color='black',label='10% background uncertainty')
plt.yscale('log')
plt.xscale('log')
plt.xticks(size=25)
plt.yticks(size=25)
plt.xlabel(r'$M_{A^{\prime}}$[GeV]',size=30, labelpad=20)
plt.ylabel(r'$\epsilon^2$',size=30, labelpad=20)
plt.minorticks_off()
plt.legend(fontsize=20)
plt.savefig(plots_dir + "sensitivity_vs_bkg_uncertainty.png")
plt.show()
```

```python
df_bdt = pd.read_csv(base_dir + "fermion_bdt_sensitivity_alpha_0.1_all_runs.csv")
df_cnn = pd.read_csv(base_dir + "fermion_CNN_sensitivity_alpha_0.1_all_runs.csv")
plt.figure(figsize=(17,14),dpi=300)
plt.plot(df_bdt['mass'],df_bdt['epsilon_squared'],color='red',label='BDT')
plt.plot(df_cnn['mass'],df_cnn['epsilon_squared'],color='blue',label='CNN')
plt.yscale('log')
plt.xscale('log')
plt.xticks(size=25)
plt.yticks(size=25)
plt.xlabel(r'$M_{A^{\prime}}$[GeV]',size=30, labelpad=20)
plt.ylabel(r'$\epsilon^2$',size=30, labelpad=20)
plt.minorticks_off()
plt.legend(fontsize=20)
plt.savefig(plots_dir + "sensitivity_bdt_vs_cnn.png")
plt.show()
```
