```python
run="run1"
base_dir = "/home/lmlepin/Desktop/dm_sets/dark_tridents_analysis/" + run + "_samples/" 
figs_dir = "/home/lmlepin/Desktop/Plots_DT_Drive/2023/truth_plots/"
df = pd.read_csv(base_dir + run + "_nu_overlay_CNN_scores_8441_steps.csv")


scalings_run1 = { "nu":1./2.35e21, "dirt":1.026*0.75/1.6e21, "beamoff": 0.98*(6230439.0/9186361.390000)}
scalings_run3 = { "nu":1./1.993661e21, "dirt":1.0*0.35/1.020e21, "beamoff": 0.98*(10522703.0/34093009.725000)}

scalings = {}

if(run == "run1"):
    run_label = "Run 1"
    target_pot = 2.2e+20
    scalings = scalings_run1
    pot_label = r'$2.2\times10^{20}$ POT'
else:
    run_label = "Run 3"
    target_pot = 5.0e20
    scalings = scalings_run3
    pot_label = r'$5.0\times10^{20}$ POT'
    
nu_scaling = target_pot*scalings["nu"]
MC_weight_branch(df)
df = df[df['signal_score'] >= 0.5]
df['signal_score'] = logit_transform(df['signal_score'])
```

```python
df_numu = df[np.abs(df['mctruth_nu_pdg']) == 14]
df_nue = df[np.abs(df['mctruth_nu_pdg']) == 12]

df_numu_cc = df_numu[df_numu['mctruth_cc_or_nc'] == 0]
df_numu_nc = df_numu[df_numu['mctruth_cc_or_nc'] == 1]
df_numu_nc_pi0 = df_numu_nc[df_numu_nc['mctruth_num_exiting_pi0'] >= 1]
df_numu_nc = df_numu_nc[df_numu_nc['mctruth_num_exiting_pi0'] == 0]


df_nue_cc = df_nue[df_nue['mctruth_cc_or_nc'] == 0]
#df_nue_cc = df_nue_cc[df_nue_cc['mctruth_num_exiting_pi0'] >= 1]
df_nue_nc = df_nue[df_nue['mctruth_cc_or_nc'] == 1]
```
