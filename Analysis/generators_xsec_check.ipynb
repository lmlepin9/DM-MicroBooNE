```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator
```

```python
df_a = pd.read_csv(input_dir + "dm_electron_xsec_mx_0.030000_bdnmc.csv")
df_b = pd.read_csv(input_dir + "dm_electron_xsec_mx_0.03_evgen.csv")
df_c = pd.read_csv(input_dir + "dm_electron_xsec_mx_0.03_evgen_Pawel_new_F.csv")


# plot globals:
axis_label_size = 45
title_size = 45
legend_text_size = 45
text_padding = 15
x_ticks_size = 35
y_ticks_size = 35
figure_size = (18,16)
```

```python
fig = plt.figure(figsize=figure_size,dpi=200)
ax = fig.add_subplot(111)

for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(2.5)  # change width
    ax.spines[axis].set_color('black')    # change color
    
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())

conversion_factor = (0.3894*1e9)*(1.9894368e-08)
#conversion_factor = 1.

plt.plot(df_a['E_val'],df_a['xsec_val']*conversion_factor,color='red',lw="3.",label="BdNMC")
plt.plot(df_c['E_val'],df_c['xsec_val']*conversion_factor/2.,color='green',lw="3.",label="DarkTridentGen")
plt.tick_params(axis="y", which='major', direction="in",length=20,width=2.5,pad=10,labelsize=40)
plt.tick_params(axis="y", which='minor', direction="in",length=10,width=1.0,labelleft=False,labelsize=40)
plt.tick_params(axis="x", which='major', direction="in",length=20,width=2.5,pad=10,labelsize=40)
plt.tick_params(axis="x", which='minor', direction="in",length=10,width=2.0,labelbottom=False,labelsize=40)
plt.yscale('log')
plt.ylim(2e-7,2e-5)
plt.grid(alpha=0.5,which='both',ls="--",lw=2.)
plt.legend(loc="center right",fontsize=45,shadow=True)
plt.title(r'Dark matter-Electron Scattering, Scalar DM',fontsize=45,pad=15)
plt.xlabel(r'Energy [GeV]',fontsize=45,labelpad=15)
plt.ylabel(r'$\sigma_{\chi e^{-}}(E_{\chi})$ [pb]',fontsize=45,labelpad=15)
plt.savefig(figs_dir + "xsec_dm_electron_comparison.pdf", bbox_inches="tight")
plt.savefig(figs_dir + "xsec_dm_electron_comparison.png", bbox_inches="tight")
plt.show()
```
