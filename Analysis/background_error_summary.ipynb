```python
import numpy as np
import matplotlib.pyplot as plt
import uproot as u
from matplotlib.ticker import AutoMinorLocator, MultipleLocator
```

```python
run="run1"
figs_dir = "/home/lmlepin/Desktop/Plots_DT_Drive/2023/multisim_var/"
base_dir = "/home/lmlepin/Desktop/dm_sets/dark_tridents_analysis/" + run + "_samples/"
collie_file = u.update(base_dir + "background_hist_{}_CNN.root".format(run))

if(run=="run1"):
    run_flag = "NuMI Run 1"
else:
    run_flag = "NuMI Run 3"
```

```python
counts_bkg, bins_bkg = collie_file['bkg_total_hist'].to_numpy()
counts_nu, bins_nu = collie_file['bkg_nu_overlay'].to_numpy()
counts_det, bins_det = collie_file['bkg_nu_DetVar_uncertainty_frac'].to_numpy()
counts_stats_nu, bins_stats_nu = collie_file['bkg_nu_overlay_stat_uncertainty'].to_numpy()
counts_stats_dirt, bins_stats_dirt = collie_file['bkg_dirt_stat_uncertainty'].to_numpy()
counts_stats_beamoff, bins_stats_beamoff = collie_file['bkg_EXT_stat_uncertainty'].to_numpy()
counts_ppfx_frac, bins_ppfx = collie_file['bkg_ppfx_uncertainty_frac'].to_numpy()
counts_genie_frac, bins_genie = collie_file['bkg_genie_uncertainty_frac'].to_numpy()
counts_reint_frac, bins_reint = collie_file['bkg_reinteraction_uncertainty_frac'].to_numpy()
```

```python
print(counts_det[0])

ppfx_error_nu = counts_nu*counts_ppfx_frac
genie_error_nu = counts_nu*counts_genie_frac
reint_error_nu = counts_nu*counts_reint_frac
det_error_nu = counts_nu*counts_det[0]
counts_stats = np.sqrt(counts_stats_nu**2 + counts_stats_dirt**2 + counts_stats_beamoff**2)
total_nu_squared = ppfx_error_nu**2 + genie_error_nu**2 + reint_error_nu**2 + counts_stats_nu**2 + det_error_nu**2

# total fractional error 
total_error = np.sqrt(total_nu_squared + counts_stats_dirt**2 + counts_stats_beamoff**2)/counts_bkg



collie_file['bkg_total_error'] = (total_error, bins_ppfx)
```

```python
print(det_error_nu)
print(counts_nu)
print(counts_bkg)
```

```python
fig = plt.figure(figsize=(20,16),dpi=300)
ax = fig.add_subplot(111)

for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(2.5)  # change width
    ax.spines[axis].set_color('black')    # change color
    
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
    
plt.hist(bins_ppfx[:-1],bins=bins_ppfx,histtype="step",weights=100*total_error,lw=4,color="black",linestyle="--",label="Quadrature sum")
plt.hist(bins_ppfx[:-1],bins=bins_ppfx,histtype="step",weights=100*counts_stats/counts_bkg,lw=4,color="red",label="Sample statistics")
plt.hist(bins_ppfx[:-1],bins=bins_ppfx,histtype="step",weights=100*det_error_nu/counts_bkg,lw=4,color="green",label="Detector")
plt.hist(bins_ppfx[:-1],bins=bins_ppfx,histtype="step",weights=100*ppfx_error_nu/counts_bkg,lw=4,color="blue",label=r"$\nu$ flux")
plt.hist(bins_genie[:-1],bins=bins_genie,histtype="step",weights=100*genie_error_nu/counts_bkg,lw=4,color="magenta",label=r"$\nu$ cross section")
plt.hist(bins_reint[:-1],bins=bins_reint,histtype="step",weights=100*reint_error_nu/counts_bkg,lw=4,color="brown",label="Reinteractions")

plt.tick_params(axis="y", which='major', direction="in",length=20,width=2.0,pad=10,labelsize=40)
plt.tick_params(axis="y", which='minor', direction="in",length=10,width=2.0,labelleft=False)
plt.tick_params(axis="x", which='major', direction="in",length=20,width=2.0,pad=10,labelsize=40)
plt.tick_params(axis="x", which='minor', direction="in",length=10,width=2.0,labelbottom=False)

plt.legend(fontsize=35,loc="upper center",ncol=2)
plt.title(r"Background Uncertainties " + run_flag,fontsize=45,pad=15)
plt.xlabel(r'CNN signal score',fontsize=45,labelpad=15)
plt.ylabel(r'% Uncertainty',fontsize=45,labelpad=15)
plt.ylim(0.,50.)
plt.xlim(0.,bins_ppfx[-1])
plt.savefig(figs_dir + run + "_background_uncertainties_summary.pdf",bbox_inches="tight")
plt.savefig(figs_dir + run + "_background_uncertainties_summary.png",bbox_inches="tight")
plt.show()
```
