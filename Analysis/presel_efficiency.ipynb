```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Efficiencies 


nu_eff = [1. , 0.41, 0.09, 0.008] 
ext_eff = [1., 0.1, 0.023, 0.003]
dirt_eff = [1., 0.24, 0.035, 0.004]
dt_eff = [1.0, 0.6888063876315923, 0.4424553730356507, 0.2577938259675533]
positions = [1, 2, 3, 4]
```

```python
plt.figure(figsize=(25,20))
plt.grid()
plt.yticks(fontsize=20)
plt.plot(positions,nu_eff,'-o',label=r'in cryo $\nu$', color = "darkviolet")
plt.plot(positions,dirt_eff,'-o',label=r'out of cryo $\nu$', color = "navy")
plt.plot(positions,ext_eff,'-o',label=r'beam-off', color = "deepskyblue")
plt.plot(positions,dt_eff,'-o',label=r'DT $M_{A} = 0.05$', color = "red")
plt.xticks([1,2,3,4],["No cut","Reco vertex", "1 shower", "0 tracks"],fontsize=20,rotation=45,ha="right")
plt.ylabel("Fraction of candidates",labelpad = 20, size=30)
plt.legend(fontsize="30")
plt.yscale('log')
plt.show()
```
