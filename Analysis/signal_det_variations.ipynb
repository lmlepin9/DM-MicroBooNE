```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
from df_utils import * 
```

```python
# Path to where you want to store your plots 
figure_path = '/home/lmlepin/Desktop/Plots_DT_Drive/2023/det_var/'

# Select a run
run = "run1"

# Path to files: 
base_dir = '/home/lmlepin/Desktop/dm_sets/dark_tridents_analysis/' + run + '_det_vars/signal/'

df_list = []
df_cv = pd.read_csv(base_dir + run + "_dt_ratio_0.6_ma_0.4_CV_CNN.csv")
df_list.append(df_cv)
print(len(df_cv))
df_cv.head()
```

```python
det_vars = ["ly_atten", "ly_down", "ly_rayleigh", "recomb", "sce", "wiremod_anglexz", "wiremod_angleyz", "wiremod_dEdx", "wiremod_x", "wiremod_yz"]

for var in det_vars:
    file_name = base_dir + run + "_dt_ratio_0.6_ma_0.4_"+var+"_CNN.csv"
    print("Opening file: " + file_name)
    df = pd.read_csv(file_name)
    print(len(df))
    df_list.append(df)

#overlap_df = make_common_evs_df(df_list)
#print(overlap_df.head())
#print(len(overlap_df))
```

```python
min_val = np.min(df_list[0]['reco_shower_energy_max'])
max_val = np.max(df_list[0]['reco_shower_energy_max'])
binning = np.linspace(0.,2000,15)


fig = plt.figure(figsize=(12, 8),dpi=300)
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.)
main_ax = fig.add_subplot(grid[:-1, :])
x_hist = fig.add_subplot(grid[-1, :], yticklabels=[-0.5,-0.25,0,0.25,0.5], sharex=main_ax)

main_ax.set_title("",size=15)
plt.xlabel('Reconstructed energy [MeV]',size=10)
main_ax.set_ylabel(r'Events',size=15)
x_hist.set_ylabel('Fractional change',size=10)


(counts_cv, bins_cv) = np.histogram(df_list[0]['reco_shower_energy_max'], bins=binning)
df_list.pop(0)
count_list = 0 

error = np.zeros(len(binning) - 1)

for df in df_list: 
    (counts, bins) = np.histogram(df['reco_shower_energy_max'], bins=binning)
    main_ax.hist(bins[:-1], bins, weights=counts,alpha=0.5, histtype='step',log=False,label=det_vars[count_list])
    diff = (counts_cv - counts)
    ratio = (counts_cv - counts) / counts_cv
    x_hist.hist(bins[:-1],bins,weights=np.nan_to_num(ratio),histtype='step',orientation='vertical',label=str(var)) 
    error=diff**2+error
    count_list+=1
main_ax.hist(bins_cv[:-1], bins_cv, weights=counts_cv, histtype='step',lw=2,label='CV',color='red',log=False)
main_ax.legend()
x_hist.set_ylim([-1.5,1.5])
#plt.savefig(figure_path + run + "_det_var_bdt_dist.png")
plt.show()
```
